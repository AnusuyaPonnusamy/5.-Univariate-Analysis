# Note:
1. Salary column has only 148 values. Mean, Median and Mode ignores NaN values in the salary column, hence returns non-null values. because skipna is true by default
2. While the percentile is considering the NaN values in the salary column
3. To ignore the NaN, 
    3.1. Pandas' quantile() which automatically ignores NaN values.
    3.2. Use NumPy's nanpercentile(), which ignores missing values.
    3.3. Use dropna() before calculating percentiles.

```python
df_MCT=pd.DataFrame(index=["Mean", "Median", "Mode","Q1:25%", "Q2:50%", "Q3:75%", "99%", "Q4:100%"], columns=quan)
for columnName in quan:
    df_MCT[columnName]["Mean"]=dataset[columnName].mean()
    df_MCT[columnName]["Median"]=dataset[columnName].median()
    df_MCT[columnName]["Mode"]=dataset[columnName].mode()[0] 
    df_MCT[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
    df_MCT[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
    df_MCT[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
    df_MCT[columnName]["99%"]=np.percentile(dataset[columnName],99)
    df_MCT[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]

df_MCT
```
