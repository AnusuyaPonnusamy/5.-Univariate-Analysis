```python
# For replacing nan values in multiple columns in the entire dataset, use the imputer function which is developed dedicatedly to handle the missing values
# Note the impute can be done only on numerical/quantitative columns
from Univariate_Library import Univariate
quan,qual=Univariate.QuanQual(dataset)
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
imp=SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
```
